```python
df = pd.read_excel('/content/drive/My Drive/IC/dados_atualizados_internacoes.xlsx')
```

```python
df.tail(10)
```

```python
df = df.iloc[:-8].reset_index(drop=True)
```

```python
df
```

```python
df = df.drop(df.columns[-1], axis=1)
```

```python
df
```

```python
# Transformação de formato largo para longo
df_long = df.melt(id_vars=['Município de residência'], var_name='Data', value_name='Internacoes')

# Exibir o DataFrame transformado
print(df_long)
```

```python
df_long['Internacoes'] = df_long['Internacoes'].astype(int)
```

```python
df_long
```

```python
df_long['Data'] = pd.to_datetime(df_long['Data'], format='%Y%m')
```

```python
df_long
```

```python
# Separando a coluna 'Município de residência' em 'Estado' e 'Cidade'
df_long[['UF', 'Município']] = df_long['Município de residência'].str.split(', ', expand=True)

# Removendo a coluna original se não for mais necessária
df_long = df_long.drop(columns=['Município de residência'])

# Exibindo o resultado
print(df_long)
```

```python
df_long = df_long.rename(columns={"Data": "Data completa", "Internacoes": "Qtd. internacoes"})
```

```python
df_long["Município"] = df_long["Município"].str.upper()
```

```python
df_long
```

```python
dataset_cidades = pd.read_csv('/content/drive/My Drive/IC/dataset_cidades.csv')
```

```python
dataset_cidades
```

```python
print(type(dataset_cidades))
```

```python
dataset_cidades["Município"] = dataset_cidades["Município"].str.upper()
```

```python
dataset_cidades
```

```python
# Realizar a junção
df_final = pd.merge(df_long, dataset_cidades, on=['Município', 'UF'], how='left')

print(df_final)
```

```python
df_final.to_csv('dados_novos.csv', index=False)
```
